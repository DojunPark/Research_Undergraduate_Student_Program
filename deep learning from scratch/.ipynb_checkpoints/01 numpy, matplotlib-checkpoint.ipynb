# nympy

```python
import numpy as np

x = np.array([1.0, 2.0, 3.0])

print(x)

print(type(x))
```

```python
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])

# sum each element
print(x + y)

# extract each element
print(x - y)

# multiplcate each element
print(x * y)

# divide each element
print(x / y)
```

```python
# n-dimensional array

A = np.array([[1, 2], [3, 4]])

print(A)
print(A.shape)
print(A.dtype)
print(A.ndim)
```

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[3, 0], [0, 6]])

print(A + B)
print(A * B)
print(A * 10)
```

```python
# broadcasting

A = np.array([[1, 2], [3, 4]])
B = np.array([10, 20])

print(A * B)
```

```python
A = np.array([[1, 2], [3, 4], [5, 6]])
B = np.array([[10, 20], [20, 30]])

print(A * B)
```

```python
X = np.array([[51, 55], [14, 19], [0, 4]])

print(X)
print(X[0])
print(X[0][1])
```

```python
for row in X:
    print(row)
```

```python
X = X.flatten()

print(X)
print(X[np.array([0, 2, 4])])
```

```python
print(X > 15)
```

```python
print(X[X>15])
```

# matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1) # create a list from 0 to 6 with interval 0.1 (not including 6!)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
print(x)
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = 'sin')
plt.plot(x, y2, linestyle='--', label = 'cos')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.image import imread

img = imread('campus.jpg')

plt.imshow(img)
plt.show()
```

```python
import matplotlib.pyplot as plt
import cv2

img = cv2.imread('campus.jpg')
plt.imshow(img)
plt.show()
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Konkuk University')
plt.axis('off')

plt.show()
```

```python
img1 = cv2.imread('campus.jpg')
img2 = cv2.imread('stanford.jpg')

fig = plt.figure(figsize = (16, 9))

ax1 = fig.add_subplot(1, 2, 1)
ax1.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
ax1.set_title('Konkuk University')
ax1.axis('off')

ax2 = fig.add_subplot(1, 2, 2)
ax2.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
ax2.set_title('Stanford University')
ax2.axis('off')

plt.show()
```
